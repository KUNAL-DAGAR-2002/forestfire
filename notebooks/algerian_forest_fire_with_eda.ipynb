```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Algerian_forest_fires_dataset.csv')
df.head(2)
```

```python
df.info()
```

```python
df.isnull().sum().sum()   #total 28 null values
```

```python
df.isnull().sum()
```

```python
df[120:]
```

```python
df.loc[:122,'region']=0
df.loc[122:,'region']=1 #Sidi-Bel Abbes Region Dataset	
df[['region']] = df[['region']].astype(int)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
#All null values are removed
```

```python
df.shape
```

```python
df.head(2)
```

```python
#Dropping day, month and year as there is no need of them
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
#Converting columns from object to int
```

```python
df[120:130]
```

```python
df.drop(124,axis=0,inplace=True)
```

```python
df = df.reset_index()
```

```python
df['Temperature'] = df['Temperature'].astype(int)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df['RH'] = df['RH'].astype(int)
df['Ws'] = df['Ws'].astype(int)
```

```python
df.info()
```

```python
df_object = df.select_dtypes('object')
```

```python
df_object.drop('Classes',axis=1,inplace=True)
```

```python
df_object.head()
```

```python
for i in df_object.columns:
    df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = df['Classes'].str.replace(' ','')
```

```python
df['Classes'].unique()
```

```python
df.Classes.value_counts()
```

```python
labels_map = {
    'fire':1,
    'notfire':0
}
```

```python
df['Classes'] = df['Classes'].map(labels_map)
```

```python
df.drop('index',axis=1,inplace=True)
```
